```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([12, 23, 31, 43, 53, 59, 63, 68, 71, 78])
y = np.array([14, 19, 25, 31, 38, 44, 49, 56, 62, 67])
z = np.array([87, 81, 73, 68, 61, 55, 49, 43, 37, 29])
```

### Line plot

```python
plt.plot(x, y);
```

```python
plt.plot(x, z);
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--', ms = 7);
plt.grid()
# markers = H, o, >, <, ^, *
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--', ms = 7);
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--', ms = 7,
        label = '2023');
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--', ms = 7,
        label = '2024');
plt.grid()
plt.legend();
```

#### subplot

```python
plt.figure(figsize=(16, 9))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--', ms = 7);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(0,90,5))
plt.yticks(range(0,90,5))
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--', ms = 7);
plt.grid()
```

```python
plt.figure(figsize=(16, 9))
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.scatter(df['Age'], df['EstimatedSalary'], 
            c = df['Purchased']);
plt.grid()
```

### Bar Plot

```python
df1 = pd.read_csv('student.csv')
df1
```

```python
plt.figure(figsize=(12,9))

plt.title('Bar Plot')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.bar(df1['name'], df1['marks'], 
        color = 'purple');
plt.grid()
```

```python
plt.figure(figsize=(12,9))

plt.title('Bar Plot')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.barh(df1['name'], df1['marks'], 
        color = 'purple');
plt.grid()
```
